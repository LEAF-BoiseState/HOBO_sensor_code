```python
import pandas as pd 
import matplotlib.pyplot as plt

hobo1_fn = '21705865 2023-07-24 17_06_41 MDT (Data MDT).csv'
hobo2_fn = '21705866 2023-07-24 17_06_48 MDT (Data MDT).csv'
```

```python
hobo1_df = pd.read_csv(hobo1_fn)
hobo1_df
```

```python
hobo2_df = pd.read_csv(hobo2_fn)
hobo2_df
```

```python
plt.figure(figsize=(14,10))
plt.rcParams.update({'font.size': 16})
plt.plot(pd.to_datetime(hobo1_df['Date-Time (MDT)']),hobo1_df['Ch:1 - Temperature   (°C)'],label='HOBO 1')
plt.plot(pd.to_datetime(hobo2_df['Date-Time (MDT)']),hobo2_df['Ch:1 - Temperature   (°C)'],label='HOBO 2')
plt.ylabel('Air Temperature [°C]')
plt.xlabel('Date')
plt.legend()
```

```python
plt.figure(figsize=(14,10))
plt.rcParams.update({'font.size': 16})
plt.plot(pd.to_datetime(hobo1_df['Date-Time (MDT)']),hobo1_df['Ch:2 - Light   (lux)'],label='HOBO 1')
plt.plot(pd.to_datetime(hobo2_df['Date-Time (MDT)']),hobo2_df['Ch:2 - Light   (lux)'],label='HOBO 2')
plt.ylabel('Light [lux]')
plt.xlabel('Date')
plt.legend()
```
